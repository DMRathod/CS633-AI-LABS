```python
import numpy as np
from copy import deepcopy
```

```python
# helper methods

def createEmptyGame():
    return np.zeros((3,3))

def emptyCells(board):
    e = 0
    for i in board:
        for j in i:
            if j==0:
                e+=1
    return e

def noMovesLeft(board):
    return emptyCells(board) == 0

def takePlayerChoice():
    print('1 for X, 2 for O, choose your move:')
    z = input()
    if z != '2' and z != '1':
        print('invalid move :\\')
        takePlayerChoice()
    else:
        return int(z)

def takeInitialMove():
    print('enter your first move. from 1 to 9')
    z = input()
    if not(z.isnumeric() and int(z)<=9):
        print('invalid move :\\')
        takeInitialMove()
    else:
        return int(z)

def setBoard(board, place, player):
    board[int((place-1)/3)][(place-1)%3] = player

def displayGame(board):
    # strange CLI asesthetics
    print('\n-------')
    for i in board:
        for j in i:
            if j == 1:
                print('X', end='|')
            elif j == 2:
                print('O', end='|')
            else:
                print('_', end='|')
        print('\n-------')
def isWon(board):
    for i in range(3):
        # check if we're winning in nth row
        if board[i][0] != 0 and board[i][0] == board[i][1] == board[i][2]:
            return board[i][0]
        # check if we're winning in nth column
        if board[0][i] != 0 and board[0][i] == board[1][i] == board[2][i]:
            return board[0][i]
    if board[0][0] != 0 and board[0][0] == board[1][1] == board[2][2]:
        return board[0][0]
    if board[2][0] != 0 and board[2][0] == board[1][1] == board[0][2]:
        return board[2][0]
    return 0

def _score(board):
    s = isWon(board)
    if s == 1:
        return 1
    elif s == 2:
        return -1
    return 0
game = [[0,1,2], [0,1,0], [2,1,2]]
# isWon(game)
```

```python
def miniMax(board, depth, isMax, move):
    if isWon(board) or noMovesLeft(board):
        return [_score(board), depth + 1]
    else:
        tempScore = -1*np.inf if isMax else np.inf
        tempDepth = np.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == 0:
                    newBoard = deepcopy(board)
                    newBoard[i][j] = move
                    newMove = 1 if move == 2 else 2
                    score, newDepth = miniMax(newBoard, depth + 1, not isMax, newMove)
                    if isMax and score>tempScore:
                        tempDepth = newDepth
                        tempScore = score
#                          and tempDepth>newDepth
                    elif not isMax and score<tempScore:
                        tempScore = score
                        tempDepth = newDepth
        return [tempScore, tempDepth]

def aiMove(board, comp):
    available = []
    minScore = -1*np.inf
    newI = newJ = 0
#     tempDepth = np.inf
    for i in range(3):
        for j in range(3):
            if board[i][j] == 0:
                newBoard = deepcopy(board)
                newBoard[i][j] = comp
                isMax = comp == 2
                player = 2 if comp == 1 else 1
                score, newDepth = miniMax(newBoard, 0, isMax, player)
#                 print('=>',i, j, score, newDepth)
                if score > minScore:
                    newI = i
                    newJ = j
                    minScore = score
#                     tempDepth = newDepth
    
    board[newI][newJ] = comp
    
def playGame():
    board = createEmptyGame()
#     board = [[2, 0, 0], [0, 0, 0], [0, 0, 0]]
    player = 2 # takePlayerChoice()
    comp = 2 if player == 1 else 1
    turn = 2
    while not(noMovesLeft(board) or isWon(board)):
        if turn == player:
            move = takeInitialMove()
            setBoard(board, move, player)
        else:
            # AI MOVES
            aiMove(board, comp)
        displayGame(board)
        turn = player if turn == comp else comp
    if isWon(board):
        if isWon(board) == player:
            print('You won!!!')
        else:
            print('You Lost!!! Hail The Bot!')
    else:
        print('Its a Tie!')
while True:
    playGame()
```
